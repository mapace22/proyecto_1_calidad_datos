Apliquemos ahora todos los cambios a la lista de clientes. Para simplificar las cosas, te proporcionaremos una más corta.

```python
users = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['CLOTHES', 'BOOKS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRONICS', 'HOME', 'FOOD'], [459, 120, 99]],
]

users_clean= []

# Procesa al primer usuario
user_name_1 = users[0][1].strip().replace('_', ' ')  # Depura espacio vacio y guión bajo
user_age_1 = int(users[0][2])  # Convierte edad (float) a entero (int)
user_name_1 = user_name_1.split()  # Divide en partes [' ', ' ', ' ',' ']
users_clean.append([users[0][0], user_name_1, user_age_1, users[0][3], users[0][4]])

# Procesa al segundo usuario
user_name_2 = users[1][1].strip().replace('_', ' ')
user_age_2 = int(users[1][2])
user_name_2 = user_name_2.split()
users_clean.append([users[1][0], user_name_2, user_age_2, users[1][3], users[1][4]])

# Procesa al tercer usuario
user_name_3 = users[2][1].strip().replace('_', ' ')
user_age_3 = int(users[2][2])
user_name_3 = user_name_3.split()
users_clean.append([users[2][0], user_name_3, user_age_3, users[2][3], users[2][4]])

print(users_clean)
print()




### INCORPORÉ OTRO EJERCICIO PARA CONVERTIR LAS MAYÚSCULAS EN minúsculas de las categorías 

users_clean_AB= []

# Procesa al primer usuario
user_name_1B = users[0][1].strip().replace('_', ' ').lower()  # Convertir a minúsculas
user_age_1B = int(users[0][2])
user_name_1B = user_name_1B.split()
categories_1B = [category.lower() for category in users[0][3]]  # Convertir a minúsculas
users_clean_AB.append([users[0][0], user_name_1B, user_age_1B, categories_1B, users[0][4]])

# Procesa al segundo usuario
user_name_2B = users[1][1].strip().replace('_', ' ').lower()
user_age_2B = int(users[1][2])
user_name_2B = user_name_2B.split()
categories_2B = [category.lower() for category in users[1][3]]
users_clean_AB.append([users[1][0], user_name_2B, user_age_2B, categories_2B, users[1][4]])

# Procesa al tercer usuario
user_name_3B = users[2][1].strip().replace('_', ' ').lower()
user_age_3B = int(users[2][2])
user_name_3B = user_name_3B.split()
categories_3B = [category.lower() for category in users[2][3]]
users_clean_AB.append([users[2][0], user_name_3B, user_age_3B, categories_3B, users[2][4]])

print(users_clean_AB)
print()



# INCORPORÉ UN TERCER EJERCICIO UNIFICANDO EN UN BUCLE LOS TRES USUARIOS CON ÍNIDCE i

users_clean_BC = []

for user in users:
    user_id = user[0]
    user_name = user[1].strip().replace('_', ' ').lower()  # Convertir a minúsculas
    user_age = int(user[2])  # Convertir la edad a entero
    user_name_split = user_name.split()  # Dividir el nombre en partes
    categories = [category.lower() for category in user[3]]  # Convertir categorías a minúsculas
    users_clean_BC.append([user_id, user_name_split, user_age, categories, user[4]])

print(users_clean_BC)

```
